```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Step 1
lifeexp = pd.read_excel('gapminder_lifeexpectancy.xlsx')
lifeexp
```

```python
# Step 2
lifeexp = lifeexp[[1950, 1975, 2000, 2015]]
lifeexp
```

```python
import matplotlib.pyplot as plt
```

```python
# Step 3: Draw a histogram for 2015 with the default settings.
lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
# Step 4: Try a few values for the number of bins.
fig = plt.figure()
ax = lifeexp[2000].hist(bins=5)
ax = lifeexp[2000].hist(bins=10)
ax = lifeexp[2000].hist(bins=20)

# Step 5: Modify the plot using some matplotlib functions.
ax.set_title('text')
ax.set_xlabel('text')
ax.set_ylabel('text')
ax
```
